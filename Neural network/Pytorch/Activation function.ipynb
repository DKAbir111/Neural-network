```python
import torch
import torch.nn as nn
```

```python
input= torch.randn(3)
input
```

# Exponential Linear Unit (ELU)

```python
fn= nn.ELU()
fn(input)
```

# Rectified linear unit (ReLU)

```python
fn= nn.ReLU()
fn(input)
```

# Leaky ReLU

```python
fn= nn.LeakyReLU(0.1)
fn(input)
```

# Sigmoid

```python
fn= nn.Sigmoid()
fn(input)
```

# Hyperbolic Tangent (Tanh)

```python
fn= nn.Tanh()
fn(input)
```

# Softmax

```python
input = torch.randn(2, 3)
fn= nn.Softmax(dim=0)
fn(input)
```

```python
fn= nn.Softmax(dim=1)
fn(input)
```

# Plotting the functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
##### Sigmoid
sigmoid = lambda x: 1 / (1 + np.exp(-x))

x=np.linspace(-10,10,10)

y=np.linspace(-10,10,100)

fig = plt.figure()
plt.plot(y,sigmoid(y),'b')

plt.grid(linestyle='--')

plt.xlabel('X Axis')

plt.ylabel('Y Axis')

plt.title('Sigmoid Function')

plt.ylim(-2, 2)
plt.xlim(-4, 4)

plt.show()
```

```python
##### TanH
tanh = lambda x: 2*sigmoid(2*x)-1

x=np.linspace(-10,10,10)

y=np.linspace(-10,10,100)

plt.plot(y,tanh(y),'b')

plt.grid(linestyle='--')

plt.xlabel('X Axis')

plt.ylabel('Y Axis')

plt.title('TanH Function')

plt.ylim(-4, 4)
plt.xlim(-4, 4)

plt.show()
```

```python
##### ReLU
relu = lambda x: np.where(x>=0, x, 0)

x=np.linspace(-10,10,10)

y=np.linspace(-10,10,1000)

plt.plot(y,relu(y),'b')

plt.grid(linestyle='--')

plt.xlabel('X Axis')

plt.ylabel('Y Axis')

plt.title('ReLU')

plt.ylim(-4, 4)
plt.xlim(-4, 4)

plt.show()
```

```python
##### Leaky ReLU
leakyrelu = lambda x: np.where(x>=0, x, 0.1*x)

x=np.linspace(-10,10,10)

y=np.linspace(-10,10,1000)

plt.plot(y,leakyrelu(y),'b')

plt.grid(linestyle='--')

plt.xlabel('X Axis')

plt.ylabel('Y Axis')

plt.title('Leaky ReLU')

plt.ylim(-4, 4)
plt.xlim(-4, 4)

plt.show()
```
