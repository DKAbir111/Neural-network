# **Import the libraries**

```python
import tensorflow as tf
import numpy as np
import math as m
import torch
import keras
import matplotlib.pyplot as plt
```

# **Sigmoid / Logistic function**

```python
xt=tf.linspace(-10,10,50)
yt=tf.keras.activations.sigmoid(xt)
plt.plot(xt,yt)
plt.title("Sigmoid function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.savefig("Sigmoid1.png")
plt.show()
```

```python
#user define function
def sigmoid_fun(xt):
  return 1/(1+ np.exp(-xt))


yt= sigmoid_fun(xt)
plt.plot(xt,yt)
plt.grid()
plt.show()
```

```python
def derivative_sigmoid_fun(xt):
  return sigmoid_fun(xt)*(1-sigmoid_fun(xt))


yt=derivative_sigmoid_fun(xt)
plt.plot(xt,yt)
plt.grid()
plt.show()
```

```python
plt.plot(xt,sigmoid_fun(xt),xt,derivative_sigmoid_fun(xt))
plt.legend(["sigmoid_xt", "derivative_sigmoid_xt"])
plt.grid()
plt.show()
```

# **TanH / Hyperbolic tangent function**

```python
xt=tf.linspace(-10,10,50)
yt=tf.keras.activations.tanh(xt)
plt.plot(xt,yt)
plt.title("tanh function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.savefig("tanh1.png")
plt.show()
```

```python
def tanh_fun(xt):
  return (np.exp(xt) - np.exp(-xt))/(np.exp(xt) + np.exp(-xt))


yt= tanh_fun(xt)
plt.plot(xt,yt)
plt.grid()
plt.show()
```

```python
def derivative_tanh_fun(xt):
  return 1 - (tanh_fun(xt)**2)
```

```python
yt=derivative_tanh_fun(xt)
plt.plot(xt,yt)
plt.grid()
plt.show()
```

```python
plt.plot(xt,tanh_fun(xt),xt,derivative_tanh_fun(xt))
plt.legend(["tanh_xt", "derivative_tanh_xt"])
plt.grid()
plt.show()
```

# **ReLU function**

```python
xt=tf.linspace(-10,10,50)
yt=tf.keras.activations.relu(xt)
plt.plot(xt,yt)
plt.title("ReLU function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.savefig("ReLU1.png")
plt.show()
```

```python
def ReLU_fun(xt):
  return np.maximum(0.0,xt)
```

```python
yt= ReLU_fun(xt)
plt.plot(xt,yt)
plt.grid()
plt.show()
```

```python
def LeakyReLU_fun(xt):
  return np.maximum(0.1*xt,xt)

yt= LeakyReLU_fun(xt)
plt.plot(xt,yt)
plt.grid()
plt.show()
```

# **Softmax Function**

```python
from scipy.special import softmax
xt=tf.linspace(-10,10,50)
yt= softmax(xt)
plt.plot(xt,yt)
plt.title("Softmax function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.savefig("Softmax1.png")
plt.show()
```
